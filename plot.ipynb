```python
import os
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas
from datetime import datetime
import time
from IPython.display import clear_output
```

```python
def plot(format, folder):
    tx_latency = []
    rx_latency = []
    for (timestamp, csv) in sorted([(datetime.strptime(f, format), f"{folder}/{f}") for f in os.listdir(folder)]):
        try:
            df = pandas.read_csv(csv)
            tx_latency.append((timestamp, df[df.columns[1]][0]))
            rx_latency.append((timestamp, df[df.columns[2]][0]))
        except:
            print(f"no data in {csv}")
    plt.figure(figsize=(20, 6))
    plt.title(f'Alpha tx and rx latency')
    plt.plot(*zip(*tx_latency), label='tx_latency')
    plt.plot(*zip(*rx_latency), label='rx_latency')
    plt.legend()
    plt.show()
```

```python
plot('%Y_%m_%d_%H_%M.csv', "./data_alpha_old")
plot('%Y_%m_%d_%H_%M_%S.csv', "./data_alpha")
# while True:
#     clear_output(wait=True)
#     plot('%Y_%m_%d_%H_%M_%S.csv', "./data_alpha")
#     time.sleep(5)
```
