```python
import os
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas
from datetime import datetime
```

```python
tx_latency = []
rx_latency = []
for (time, csv) in sorted([(datetime.strptime(f, '%Y_%m_%d_%H_%M.csv'), f"./data_alpha/{f}") for f in os.listdir("./data_alpha")]):
    df = pandas.read_csv(csv)
    tx_latency.append((time, df[df.columns[1]][0]))
    rx_latency.append((time, df[df.columns[2]][0]))
```

```python
plt.figure(figsize=(20, 6))
plt.title(f'Alpha tx and rx latency')
plt.plot(*zip(*tx_latency), label='tx_latency')
plt.plot(*zip(*rx_latency), label='rx_latency')
plt.legend()
plt.show()
```
